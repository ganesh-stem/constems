# **1. Generate the Random Values.**

```python
# initilizing the essential values
first_list_of_random_values = []
second_list_of_random_values = []
epochs = 100000 # number of iterations
```

```python
# Defining the function to get the pseudo-random numbers.
# m (> 0) stands for modulus parameter, a of range (0, m) is multiplier,
# c of range (0, m) is a incrementer
# linearCongruentialMethod.seed_value is the sequence of pseudo-random numbers. 

def linearCongruentialMethod(m = 2**32, a = 1103515245, c = 12345):
    linearCongruentialMethod.seed_value = (a * linearCongruentialMethod.seed_value + c) % m 
    return linearCongruentialMethod.seed_value / m
```

```python
# "linearCongruentialMethod.seed_value = 1" is Initial value of the sequence known as seed.
linearCongruentialMethod.seed_value = 1 

for i in range(epochs): 
    pseudo_random = linearCongruentialMethod()
    first_list_of_random_values.append(pseudo_random)
```

```python
# "linearCongruentialMethod.seed_value = 2" is Initial value of sequence known as seed.
linearCongruentialMethod.seed_value = 2
 
for i in range(epochs): 
    pseudo_random = linearCongruentialMethod()
    second_list_of_random_values.append(pseudo_random)
```

# **2. Visualizing the Random Values**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn.categorical import violinplot
```

```python
def kde_and_displot(x_label, inputValue):
  ax = sns.kdeplot(inputValue, shade = True);
  ax.set(xlabel= x_label, ylabel='DENSITIES', title=f'{x_label} VS DENSITIES');

  ax = sns.displot(inputValue);
  ax.set(xlabel=x_label, ylabel='FREQUENCIES', title=f'{x_label} VS FREQUENCIES');

def violin_and_displot(axes_label, inputValue):
  plt.figure(figsize=(20,300))
  plt.subplot(60,3,1)
  plt.title(f"{axes_label} with their DENSITIES and the MEDIAN VALUE.")
  plt.xlabel("DENSITIES")
  plt.ylabel(axes_label)
  plt.violinplot(inputValue, showmedians = True);
  
  plt.figure(figsize=(20,300))
  plt.subplot(60,3,2)
  plt.title(f"{axes_label} and the MEAN VALUE.")
  plt.ylabel(axes_label)
  plt.boxplot(inputValue);
```

```python
kde_and_displot('RANDOM VALUES', first_list_of_random_values)
```

```python
violin_and_displot("RANDOM VALUE", first_list_of_random_values)
```

```python
kde_and_displot('RANDOM VALUES', second_list_of_random_values)
```

```python
violin_and_displot("RANDOM VALUE", second_list_of_random_values)
```

# **3. Predict the value of Pi upto two decimal places.**

```python
def piValues(firstList, secondList):
  #  Initilizing total point generated and points withing the circle to zero.
  total_points = 0
  within_circle = 0
  pi_values = []
  for i in range(epochs):
    X = firstList[i]
    Y = secondList[i]

    total_points += 1

    distance = (X**2) + (Y**2)
    distance = (distance ** 0.5)
    
    if distance < 1:
        within_circle += 1
    if total_points % 1000 == 0:
        pi_estimate = 4 * within_circle / total_points
        pi_values.append(pi_estimate)
  return pi_values
```

```python
list_of_pi_values = piValues(first_list_of_random_values, second_list_of_random_values)
```
